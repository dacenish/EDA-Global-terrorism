The Global Terrorism Database (GTD) is an open-source database including information on terrorist attacks around the world from 1970 through 2017. The GTD includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period and now includes more than 180,000 attacks. The database is maintained by researchers at the National Consortium for the Study of Terrorism and Responses to Terrorism (START), headquartered at the University of Maryland.

# **5. Casualty Analysis.**

##**5.1 Demography of casualities by location.**


```python
# Prepare dataframes
top_countries = gt_df.groupby('Country')['Kills'].sum().nlargest(10)
top_countries = top_countries.drop('Unknown', errors='ignore')

top_regions = gt_df.groupby('Region')['Kills'].sum().nlargest(10)
top_regions = top_regions.drop('Unknown', errors='ignore')

top_states = gt_df.groupby('State')['Kills'].sum().nlargest(11)
top_states = top_states.drop('Unknown', errors='ignore')

top_cities = gt_df.groupby('City')['Kills'].sum().nlargest(11)
top_cities = top_cities.drop('Unknown', errors='ignore')

# Plot Subplot of Countries and add respective lables and tiltes 
plt.subplot(2,2,1)
sns.barplot(y=top_countries.index, x=top_countries.values, palette = 'Blues', orient = 'h')
plt.xlabel('Country')
plt.ylabel('Kills')
plt.title('Top 10 countries with the highest casualties', fontsize=15, fontweight='bold')
plt.xticks(rotation = 45)

# Plot Subplot of States and add respective lables and tiltes
plt.subplot(2,2,2)
sns.barplot(y=top_states.index, x=top_states.values, palette = 'crest', orient = 'h')
plt.xlabel('State')
plt.ylabel('Kills')
plt.title('Top 10 States with the highest casualties')
plt.xticks(rotation=45)

# Plot Subplot of Regions and add respective lables and tiltes
plt.subplot(2,2,3)
sns.barplot(y=top_regions.index, x=top_regions.values, palette = 'GnBu_d', orient = 'h')
plt.xlabel('Region')
plt.ylabel('Kills')
plt.title('Top 10 Regions with the highest casualties')
plt.xticks(rotation=45)

# Plot Subplot of Cities and add respective lables and tiltes
plt.subplot(2,2,4)
sns.barplot(y=top_cities.index, x=top_cities.values, palette = 'flare', orient = 'h')
plt.xlabel('City')
plt.ylabel('Kills')
plt.title('Top 10 Cities with the highest casualties')
plt.xticks(rotation=45)

# Set plot size
plt.gcf().set_size_inches(15,15)

# Display plot
plt.show()
```

**Observation:**


1) Among the top 10 countries, **Iqar** has the **most casualties**, while **El Salvador** has the **fewest**.

2) Among the top 10 states, **Bugdab** has the **most casualties**, while **Punjab(India)** has the **fewest**.

3) Among the top 10 regions, the **Middle East and North Africa** have the **highest number** of casualties, while **East Asia** has the **fewest**.

4) Among the top 10 cities, **Bagdad and North Africa** have the **highest number of casualties**, while **Maiduguri** has the **lowest**.


##**5.2 Demography of casualities by Terrorist Groups**


```python
# Prepare datafrme
most_kills = gt_df.groupby('Group_name')['Kills'].sum().nlargest(6)
most_kills = most_kills.drop('Unknown')

# Plot Pieplot
plt.figure(figsize = (7,7))
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff6699']
plt.pie(most_kills, labels = most_kills.sort_values().index, autopct='%1.1f%%',
        wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' }, colors = colors, pctdistance= 0.7)

my_circle = plt.Circle((0,0), 0.5, color = 'white')
plt.gcf().gca().add_artist(my_circle)

# Add title
plt.title('Top 5 terrorist groups with the highest casualties')
plt.xlabel('Percentage Of kills', fontsize=12)
```

**Observation:**

1) Among the top 5 terrorist groups **Lineration Tigers of Tamil Eelam(LTTE)** caused the **most** casualties of **35.0 percentage**. :

2) Among the top 5 terrorist groups **Islamic State of Iraq the Levant(ISIL)** caused the **least** casualties of **9.9 percentage**.



##**5.2 Demography of casualities by year**


```python

year_kills = gt_df.groupby('Year')['Kills'].sum()
year_kills = year_kills.drop('Unknown', errors = 'ignore') 

# Plot the KDE plot
plt.figure(figsize=(7, 7))
sns.kdeplot(data=year_kills, x=year_kills.index, y=year_kills.values, fill=True, cmap='Blues')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Density')
plt.title('Density Plot: Total Kills by Year')


# Show the plot
plt.show()
```

**Observation:**

1) The density of kills was relatively **low** in the **early years (1970s)** but started to increase gradually.

2) There was a significant **increase in the density** of kills from the **1980s** to the **early 1990s**.

3)The density of **kills reached a peak around the mid-1990s** to early 2000s and then started to fluctuate.

4)From the **mid-2000s** onwards, there was a notable **increase in the density of kills**, with some fluctuations in between.

5)The **highest density** of kills occurred in the years **2014, 2015, and 2016**, indicating a period of increased violence.
